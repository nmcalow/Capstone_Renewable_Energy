## Hydroelectric Power: [Source](https://en.wikipedia.org/wiki/Hydroelectric_power_in_the_United_States)

- Hydropower is weird. There are two aspects to it, hydroelectic energy and hydroelectric pumped storage, which uses excess energy to pump water back into the reservoir so it can be used to generate more electricity, converting electrical energy into potential.
- It is also heavily reliant on annual rainfall, and as such is incredibly seasonal in all sorts of fun, unexpected ways

## Key Takeaways

Hydro is incredibly seasonal, but is still following a definite trend. There are only 11 hydroelectric plants currently in the United States, with no concrete plans to build more any time soon. That being said, there are incremental increases in plant efficiency that should marginally increase hydropower in the United states. Additionally, they are not being used at 100% capacity, or even 50% capacity, so more could be utilized in the future.

## Hydro is weird enough. Let's see how RNNs work for it

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from keras.models import Sequential
from keras.layers import Dense, Dropout, GRU, LSTM
from keras.optimizers import Adam
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
hydro.drop('cap', axis = 1, inplace = True)
hydro.set_index('ds', inplace = True)
hydro.head()
```

```python
hydro.plot(figsize = (20,12))
```

```python
X = np.array(hydro.index).reshape(-1,1)

y = np.array(hydro['y']).reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,shuffle = False, test_size = 0.25, random_state = 42)
```

```python
train_generator = TimeseriesGenerator(X_train, y_train, length = 12, batch_size = 1)
test_generator = TimeseriesGenerator(X_test, y_test, length = 12, batch_size = 1)
```

```python
drop_rate = 0.5
model = Sequential()
model.add(LSTM(24, return_sequences = False, dropout = drop_rate, recurrent_dropout = drop_rate, input_shape = (12,1)))
#model.add(LSTM(128, return_sequences = False, dropout = drop_rate, recurrent_dropout = drop_rate))

model.add(Dense(16, activation = 'relu'))
model.add(Dropout(rate = drop_rate))
model.add(Dense(8, activation = 'relu'))
model.add(Dropout(rate = drop_rate))

# model.add(Dense(16, activation = 'relu'))
# model.add(Dropout(rate = drop_rate))
# model.add(Dense(16, activation = 'relu'))
# model.add(Dropout(rate = drop_rate))
# model.add(Dense(16, activation = 'relu'))
# model.add(Dropout(rate = drop_rate))

model.add(Dense(1, activation = 'linear'))

model.compile(loss = 'mean_squared_error', optimizer = Adam(lr = 0.035))
history = model.fit_generator(train_generator, epochs = 100, validation_data = test_generator, verbose = 2)
```

```python
plt.figure(figsize = (20,12))
plt.plot(history.history['loss'], label = 'Train')
plt.plot(history.history['val_loss'], label = "Test")
plt.legend()
```

```python
pred_train = model.predict_generator(train_generator)

plt.figure(figsize = (20,12))
plt.plot(pred_train, label = 'Prediction')
plt.plot(y_train, label = 'Actual')
plt.legend()
```

```python
pred_test = model.predict_generator(test_generator)

plt.figure(figsize = (20,12))
plt.plot(pred_test, label = 'Prediction')
plt.plot(y_test, label = 'Actual')
plt.legend()
```
