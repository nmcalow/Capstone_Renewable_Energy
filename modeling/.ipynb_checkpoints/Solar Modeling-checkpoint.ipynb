## Solar or Photovoltaic energy: [Source](https://www.seia.org/us-solar-market-insight)

- Solar energy takes energy from sunlight and transmits it into electricity through phosovoltaic panels in fields, on rooftops, or in large, purpose-built facilities. 
- The major trouble with solar has to do with its unreliability, as cloudy days, seasonal changes, and precipitation greatly impact the total energy generated that day
- Another hurdle to cross is the availiability of electricity storage, whether by use of large batteries, or other, indirect methods of power smoothing(Hydroelectric pumped storage, for example)
