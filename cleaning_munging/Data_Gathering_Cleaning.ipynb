## Getting the dates to be the index and converting them to proper Datetime form

```python
df_recent.index
```

```python
abv_month = {
    'Jan': '01',
    'Feb': '02',
    'Mar': '03',
    'Apr': '04',
    'May': '05',
    'Jun': '06',
    'Jul': '07',
    'Aug': '08',
    'Sep': '09',
    'Oct': '10',
    'Nov': '11',
    'Dec': '12'
}
```

```python
def getDate_recent(month_year):
    month, year = month_year.split()
    month = abv_month[month]
    return datetime.date.fromisoformat(f'{year}-{month}-01')
```

```python
def getDate_past(year):
    return datetime.date.fromisoformat(f'{year}-12-01')
```

```python
index_before_past = df_past.index
index_after_past = []
for item in index_before_past:
    index_after_past.append(getDate_past(int(item)))
    
```

```python
df_past['date'] = index_after_past
df_past.set_index('date', inplace = True)
```

```python
df_past.columns
```

```python
index_before_recent = df_recent.index
index_after_recent = []
for item in index_before_recent:
    index_after_recent.append(getDate_recent(item))
    
```

```python
df_recent['date'] = index_after_recent
```

```python
df_recent.set_index("date", inplace = True)
```
