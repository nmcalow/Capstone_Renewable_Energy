```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
```

```python
df_recent = pd.read_csv('../data/energy_recent_clean.csv', index_col = 'date')
df_past = pd.read_csv('../data/energy_past_clean.csv', index_col = 'date')
```

```python
df_recent.set_index(pd.to_datetime(df_recent.index), inplace = True)
df_past.set_index(pd.to_datetime(df_past.index), inplace = True)
```

```python
features_past = [
    'Conventional Hydroelectric Power', 
    'Wind',
    'Geo- thermal', 
    'Solar/PV', 
    'Renewable Total',
    'Renewable Percentage'
]
```

```python
features = [
    'Conventional Hydroelectric Power', 
    'Wind',
    'Geo- thermal', 
    'Solar/PV', 
    'Biomass Total'
]
```

```python
df_recent[features].tail()
```

```python
df_past[features_past].head()
```

```python
plt.figure(figsize = (12,12))
sns.heatmap(df_recent[features].corr(), cmap = 'coolwarm', vmin = -1, annot = True )
```

```python
plt.figure(figsize = (12,12))
sns.pairplot(df_recent[features])
```

## Hydroelectric

```python
df_recent['Conventional Hydroelectric Power'].describe()
```

```python
df_recent['Conventional Hydroelectric Power'].plot(figsize = (20,12))
df_past['Conventional Hydroelectric Power'].plot(color = 'green')
```

```python
plt.figure(figsize = (20,12))
df_recent['Conventional Hydroelectric Power'].plot()
df_recent['Conventional Hydroelectric Power'].rolling(12).mean().plot(color = 'red')
df_recent['Conventional Hydroelectric Power'].rolling(24).mean().plot(color = 'green')
df_recent['Conventional Hydroelectric Power'].rolling(36).mean().plot(color = 'black')
df_recent['Conventional Hydroelectric Power'].rolling(48).mean().plot(color = 'purple')
plt.axhline(22504, color = 'red')
```

```python
plt.figure(figsize = (20,12))
df_recent['Conventional Hydroelectric Power'].diff().plot()
plt.axhline(df_recent['Conventional Hydroelectric Power'].diff().mean(), color = 'red')
```

```python
decomp = seasonal_decompose(df_recent['Conventional Hydroelectric Power'])
decomp.plot();
```

```python
plot_acf(df_recent['Conventional Hydroelectric Power'], lags = 96);
```

```python
plot_pacf(df_recent['Conventional Hydroelectric Power'], lags = 72);
```

## Initial Observations:

- Solar only really started in 2014 and has since skyrocketed. What is the rate of new solar installations? What is the rate of efficiency?
- Wind is a strong contender as well. What is the rate of new wind projects? How quickly is effeciency increasing?
- Biomass and Geothermal are relative footnotes comparatively
- Hydroelectric is very stable, but with some very interesting seasonality that has the most to do with global weather patterns, drought, etc. How many new hydroelectric projects are coming up?
