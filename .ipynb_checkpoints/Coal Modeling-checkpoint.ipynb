# SARIMAX Modeling

```python
df_coal = pd.concat([df_recent[['Coal']], df_past[['Coal']]])
```

```python
df_coal.head()
```

```python
df_coal.head()
```

```python
df_recent['first_coal'] = df_recent['Coal'].diff()
df_recent['second_coal'] = df_recent['Coal'].diff().diff()
```

```python
df_coal['first_coal'] = df_coal['Coal'].diff()
df_coal['second_coal'] = df_coal['Coal'].diff().diff()
```

### Just Recent

```python
df_recent['first_coal'].mean()
```

```python
adfuller(df_recent['Coal'])
```

```python
adfuller(df_recent['first_coal'].dropna())
```

```python
train, test = train_test_split(df_recent['first_coal'].dropna(), shuffle = False)
```

```python
test.tail()
```

```python
list(range(0,10))
```

```python
# Fit a SARIMA(0, 2, 1) x (1, 0, 0, 12) model.

sarima = SARIMAX(endog = train,
                 order = (7,1,0),              # (p, d, q)
                 seasonal_order = (6,1,0,12)) # (P, D, Q, S)

# Fit SARIMA model.
model = sarima.fit()

# Generate predictions based on test set.
preds = model.predict(start=datetime(2014,10,1),
                      end=datetime(2019,4,1))

# Evaluate predictions.
print(mean_squared_error(test, preds))

# Plot data.
plt.figure(figsize=(20,12))
plt.plot(train, color = 'blue')
plt.plot(test, color = 'orange')
plt.plot(preds, color = 'green')
plt.title(label = 'Coal Electricity Generation with SARIMA(6,1,0) x (6, 1, 0, 12) Predictions', fontsize=16)
plt.legend()
plt.show();
```
